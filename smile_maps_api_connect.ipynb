## Reading in clinic location and restoration data

```python
clinic_locat = pd.read_csv('data/clinic_restoration_tofxn.csv')
clinic_locat
```

## Adding web address column for creating the base url

```python
clinic_locat['web_address'] = ''
clinic_locat
```

## Using for loop to add '+' for spaces for base url

```python
for index, row in clinic_locat.iterrows():
    a = row['Full Address']
    clinic_locat.loc[index, 'web_address'] = a.replace(' ','+')
clinic_locat
# .loc allows for you to zero in on certain cells
```

```python
address_3 = clinic_locat['web_address']
```

## Creating for loop to reach out to API and grab the addresses for each individual clinic

```python
geo_code = []
geo_code
# creating empty dataframe to store addresses
```

```python

for index, row in clinic_locat.iterrows():
    geo_code # reference created df
    d = row['Full Address'] # assigning variable to address column
    params = {'address' : d,
                'key' : api_key} # defining the paramaters required by API
    resp = requests.get(endpoint, params = params) # getting a response from API
    result = resp.json() # storing response as json
    geo_code1 = result['results'][0]['geometry']['location'] # pulling the geo coordinates and storing in variable
    geo_code2 = pd.DataFrame(geo_code1, index=[0]) # converting list to df
    print(geo_code1) # print for test to check if pulling correct values
    geo_code.append(geo_code1) # appending desired results from API into the previously created df
    time.sleep(5) # 5 sec pings to API to not overload or go over the limit/sec
geo_code
```

```python
geo_code3 = pd.DataFrame(geo_code)
```

```python
type(geo_code3)
```

```python
geo_code3['Address'] = clinic_locat['Full Address']
geo_code3 # attaching the addresses to each lat lng for merging later
```

## Creating merged df

```python
clinics = pd.merge(clinic_locat, geo_code3, left_on = 'Full Address', right_on = 'Address')
clinics = clinics.drop(['web_address', 'Address'], axis = 1)
clinics 
```

```python
clinics['Percent of Successful Restoration of Function'] = clinics['Percent of Successful Restoration of Function'].astype(float).round(2)
clinics
```

```python
clinics.to_csv(r'data\clinics_geo.csv', index = False)
```

## Creating radius map using above df

```python
area_center = [36.1627, -86.7816]
map_clinic = folium.Map(location =  area_center, zoom_start = 7)
#convert to float, radius arg seems to only work that way
clinics['Total Number of Times Care Provided'] = clinics['Total Number of Times Care Provided'].astype(float)
for row_index, row_values in clinics.iterrows():
    loc = [row_values['lat'], row_values['lng']] # references lat and lng columns
    pop = row_values['Site Info'] + '<br>' + 'Percent of Successful Restoration of Function per Procedure' + ' ' + str(row_values['Percent of Successful Restoration of Function']) +'%'
    marker = folium.Circle(
        location = loc, 
        popup = pop, 
        color = 'red',
        fill = True,
        fill_color = 'red',
        radius = clinics.iloc[row_index]['Total Number of Times Care Provided']*5).add_to(map_clinic)
map_clinic
```
