## Imports

```python
import requests 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Read in Excel File

```python
smile_data = pd.read_excel('data/SmileOnexcel.xlsx', skiprows = 2,  usecols = ['CallerNum',
                                                                               'SMILE ON 60+ Screening - Last Screening Date', 
                                                                               'SMILE ON 60+ Oral Care Encounter - What care was provided to the enrollee? (Select all that apply)',
                                                                              'SMILE ON 60+ Site Information - Enroller ID',
                                                                               'CityName',
                                                                               'CountyName',
                                                                               'PostalCode',
                                                                              'SMILE ON 60+ Treatment Plan - Is the treatment plan completed?',
                                                                              'SMILE ON 60+ Treatment Plan - Was a treatment plan developed?',
                                                                              'SMILE ON 60+ Treatment Plan - Was the enrollee able to be restored to function (can chew) and "social six" esthetics (top front six teeth are present and disease free)'])
smile_data.tail(20)
```

```python
smile_data.info()
```

## Rename Columns

```python
smile_data.columns = ['ID Number',
                      'City Name',
                      'County',
                      'Zip',
                      'Last Screening Date', 
                      'Care Provided', 
                      'Site Info', 
                      'Treatment Plan Completed?', 
                      'Treatment Plan Developed',
                     'Restored to Fxn?']
smile_data.head()
```

## Examine Negative values in ID column

```python
smile_data_negatives = smile_data[smile_data['ID Number'] < 0]
smile_data_negatives
```

## Select only Nonnegative Values in ID column

```python
smile_data2 = smile_data[smile_data['ID Number'] > 0]
smile_data2
```

```python
smile_sites = smile_data2.groupby(['Site Info']).agg(['count'])
smile_sites = smile_sites.drop(['City Name', 'County', 'Zip', 'Last Screening Date','Care Provided', 'Treatment Plan Completed?', 'Treatment Plan Developed', 'Restored to Fxn?'], axis=1)
smile_sites
```

## Select only rows where there was restored fxn

```python
smile_data_restoredfxn = smile_data2[smile_data2['Restored to Fxn?'] == 'Yes']
smile_data_restoredfxn
```

## Select only rows where there was nonrestored fxn

```python
smile_data_nonrestoredfxn = smile_data2[smile_data2['Restored to Fxn?'] == 'No']
smile_data_nonrestoredfxn
```

## Look at sites that had the most success restoring fxn

```python
smile_rfxn_site = smile_data_restoredfxn.groupby(['Site Info']).agg(['count'])
smile_rfxn_site
```

```python
successful_site = smile_rfxn_site.drop(['ID Number', 'City Name', 'County', 'Zip', 'Last Screening Date','Care Provided', 'Treatment Plan Completed?', 'Treatment Plan Developed'], axis=1)
successful_site
```

## Look at sites that had the least success restoring fxn

```python
smile_nrfxn_site = smile_data_nonrestoredfxn.groupby(['Site Info']).agg(['count'])
smile_nrfxn_site
```

```python
unsuccessful_site = smile_nrfxn_site.drop(['ID Number', 'City Name', 'County', 'Zip', 'Last Screening Date','Care Provided', 'Treatment Plan Completed?', 'Treatment Plan Developed'], axis=1)
unsuccessful_site
```

## Combining 2 tables success and unsuccessful

```python
clinics = pd.merge(successful_site, unsuccessful_site, left_on = 'Site Info', right_on = 'Site Info')
clinics.columns = ['Successful Function Restoration', 'Unsuccessful Function Restoration']
clinics = pd.merge(clinics, smile_sites, left_on = 'Site Info', right_on = 'Site Info')
clinics.columns = ['Successful Function Restoration', 'Unsuccessful Function Restoration', 'Total Number of Times Care Provided']
clinics['Percent Successful Procedures'] = (clinics['Successful Function Restoration']/clinics['Total Number of Times Care Provided'])*100
clinics
```

## Finding Duplicates for people that had multiple procedures or visits

```python
smile_dupes = pd.concat(g for _, 
              g in smile_data2.groupby("ID Number") 
              if len(g) > 1)
smile_dupes
```

## People that had both multiple procedures/visits and completed a treatment plan

```python
smile_dupes_comp_treat = smile_dupes[smile_dupes['Treatment Plan Completed?'] == 'Yes']
smile_dupes_comp_treat.head(50)
```

## Unique People that have had multiple procedures or visits

```python
smile_dupes_unique = smile_dupes.drop_duplicates(subset=['ID Number'])
smile_dupes_unique
```
